```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName('Read File txt')
sc = SparkContext.getOrCreate(conf=conf)
rdd = sc.textFile('alice.txt')
#rdd.collect()

rdd2 = rdd.flatMap(lambda x: x.split(' '))
#rdd2.collect()

def get_word_starting_with_vow(word):
    if word != '' and word[0].lower() in 'aeiou':
        return True
    else:
        return False
rdd3 = rdd2.filter(get_word_starting_with_vow)
rdd3.collect()
```
