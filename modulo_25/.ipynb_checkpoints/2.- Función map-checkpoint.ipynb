```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName('Read File txt')
sc = SparkContext.getOrCreate(conf=conf)
rdd = sc.textFile('alice.txt')
rdd.collect()
```

```python
rdd2 = rdd.map(lambda x: x.split(' '))
rdd2.collect()
```

```python
def find_length_of_words(text):
    words = text.split(' ')
    words_and_length = []
    for word in words:
        words_and_length.append({word: len(word)})
    return words_and_length
rdd3 = rdd.map(find_length_of_words)
```

```python
rdd3.collect()
```
