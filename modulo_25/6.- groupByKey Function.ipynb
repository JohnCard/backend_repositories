```python
from pyspark import SparkConf, SparkContext
import re

regex = re.compile('[^a-zA-Z]')

conf = SparkConf().setAppName('groupbyKey Function')
sc = SparkContext.getOrCreate(conf=conf)
rdd = sc.textFile('alice.txt')
rdd.collect()

rdd2 = rdd.flatMap(lambda x: x.split(' ')).map(lambda x: x.lower()).distinct()
rdd3 = rdd2.map(lambda x: (len (x),x))
rdd3.collect()

rdd3.groupByKey().mapValues(list).collect()
```
